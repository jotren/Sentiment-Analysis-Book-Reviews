# Result Analysis

After processing we now need to see whether the sentiment analysis is correct with spot checking and analytics.

```python
import pandas as pd
```

```python
df_original = pd.read_csv("../data/raw/Books_rating.csv")
```

```python
df_sentiment = pd.read_csv("../data/result/Books_summary_with_sentiment.csv")
```

```python
# Select the first 10,000 rows from the original DataFrame
df_original = df_original.iloc[:10000]

# Drop the 'Id' column from the sentiment DataFrame
df_sentiment = df_sentiment.drop(columns=['Id'])

# Merge the two DataFrames
# Assuming you want to merge them based on their indices (row numbers)
df_merged = pd.merge(df_original, df_sentiment, left_index=True, right_index=True)
```

```python
df_merged.head()
```

```python
negative = df_merged[df_merged['sentiment_label'] == 'NEGATIVE']
```

```python
pd.set_option('display.max_colwidth', None)

negative_only = negative[['review/text','summary', 'sentiment_score']].head(-5)
```

```python
print(len(negative_only[negative_only['sentiment_score'] < 0.7]))
```
